###  To fix this issue You will fill these values with ........... for numeric and ......for categorical

```python
# trying to fill missing values in my datset using for loop 

for col in df.columns:
    if df[col].dtype in ['int64', 'float64']:
        df[col].fillna(df[col].median(), inplace=True)
    else:
        df[col].fillna(df[col].mode()[0], inplace=True)
```

```python
##  step  2 ) Incorrect data type as in (( `user_type` &&	`member_gender`)) :  significant change , so  need to clean it.
df['user_type'] = (
    df['user_type']
    .str.strip()
    .str.lower()
    .replace({
        'subscriber': 'Subscriber',
        'customer': 'Customer'
    })
)
```

```python
df['member_gender'] = (
    df['member_gender']
    .str.strip()
    .str.lower()
    .replace({
        'male': 'Male',
        'female': 'Female',
        'other': 'Unknown'
    })
)
```

```python
## check your work 
print(df[['user_type', 'member_gender']].nunique())
print(df[['user_type', 'member_gender']].value_counts())
```

```python
## step 3 ) Good factor for marketing work as in ((`member_birth_year`)) : so need to clean it.
## To know the target audience's age and serve on them

# create a column for users' age depending on the current year
df['member_birth_year'] = pd.to_numeric(df['member_birth_year'], errors='coerce')

current_year = 2019
df = df[
    (df['member_birth_year'] >= current_year - 80) &
    (df['member_birth_year'] <= current_year - 18)
]

df['age'] = current_year - df['member_birth_year']
```

```python
## check your work 
df[['member_birth_year', 'age']].head()
```

```python
## change the new column age data type from float to int
df['age'] = df['age'].astype(int)
```

```python
df['age'].describe()
```

```python
## create a function to know the average users ages
def average_user_age(df, age_col='age'):
    return round(df[age_col].mean(), 2)
```

```python
avg_age = average_user_age(df)
avg_age
```

```python
## another important factor to offer  marketing services is the trip duration time 

## need to change the time from second to minute to be more usable
df['duration_min'] = (df['duration_sec'] / 60).round(2)
```

```python
## I need to change the time from second to Houre  to be more usable
df['duration_hour'] = (df['duration_sec'] / 3600).round(2)
```
